## Agregar enlaces en la zona de custom_urls

Si desea agregar enlaces en la zona de `custom_urls`, debe seguir estas reglas:

1. Identifique el tipo de enlace si es un LORA, modelo o TextualInversion.
2. Asigne una etiqueta antes del enlace.
Hashtags validos: `#embed`, `#model`, `#hypernet`, `#lora`, `#lyco`, `#vae`,`#addnetlora`
3. Agregue la etiqueta y el enlace como se indica en el ejemplo: `#model <enlace de Civitai>`.

Ejemplo:

```
 #lora https://civitai.com/api/download/models/6114,#model lhttps://civitai.com/api/download/models/25526,
```
